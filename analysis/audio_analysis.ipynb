```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
# Load audio file
FILE_NAME = 'sine.mp3'
audio_file = FILE_NAME
audio, sr = librosa.load(audio_file)
```

```python
# Display Waveform
librosa.display.waveshow(audio, sr=sr)
```

```python
# Extract MFCCs
mfccs = librosa.feature.mfcc(y=audio, sr=sr)
# Display MFCCs
librosa.display.specshow(mfccs, x_axis='time')
```

```python
fourier = librosa.stft(audio)
s = librosa.amplitude_to_db(abs(fourier))
librosa.display.specshow(s, sr=sr, x_axis = 'time', y_axis='linear')
```

```python
chroma = librosa.feature.chroma_stft(S=fourier, sr=sr)
librosa.display.specshow(chroma, sr=sr)
```
